```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
############ Basic Graph
```

```python
X=[0,1,2,3,4]
Y=[0,2,4,6,8]

# resize your Graph
plt.figure(figsize=(6,4), dpi=300)

plt.plot(X,Y, label='try', color='blue', linewidth=4, marker='o', linestyle='--',markersize=10, markeredgecolor='blue')

# line Number 2
a= np.arange(0,4.5,0.5)
b= a**2
# plt.plot(a,b, label='try02', color='y', linewidth=2, marker='^', linestyle=':',markersize=10, markeredgecolor='red')
plt.plot(a[:6],b[:6],'r', label='try02')
plt.plot(a[5:],b[5:],'r--')

print(a)
print(a[:6])
print(a[5:])

# fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20} to change font and size
plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname': 'Arial', 'fontsize': 20})
plt.ylabel('Y Axis', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# use ticks([xxx]) to set the tick marks
plt.xticks([0,1,2,3,4])
# plt.yticks(np.arange(0,150,15))

plt.legend()

# save this graph
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

```python
### this one cell is teacher's example

x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

```python
# read data
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

```python
## use this data for Histograms
```

```python
bins = [40,50,60,70,80,90,100]

## use google search "color picker" find the color you like and paste the hex code 
plt.hist(fifa['Overall'], bins=bins, color='#fc0377')

plt.xticks(bins)

plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

```python
#### for pie chart
```

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] =='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] =='Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#216beb', '#ff3b3b']
## use autopct='%.2f %%' to add number in this pie chart, %.2f means to show 2  decimals, and %% shows percentages  
plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
fifa.Weight
```

```python
## strip('lbs') to get rid of 'lbs'
## x.strip('lbs') if type(x)==str   means if the values are string, then get rid of 'lbs'
## else x for x in fifa.Weight      menas otherwise just leave x as it is(just don't change) 
## use int() function to change fifa.Weight from string to integer
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.figure(figsize=(8,5))
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weight = [light, light_medium, medium, medium_heavy, heavy]
labels = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4, .1, 0, 0, .4)

plt.title('Weight of Professional Soccer Players (lbs)')

## pctdistance= 1.2 to set the numbers how far from the center
plt.pie(weight, labels=labels,autopct='%.2f %%', pctdistance= 0.8, explode=explode)


plt.show()
```

```python
########### box chart
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8), dpi=100)

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

labels=['FC Barcelona','Real Madrid','NE Revolution']

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

## patch_artist=True so that  box.set(facecolor = '#e0e0e0' ) can be use
## medianprops={'linewidth':2} change the linewidth of each box's median line
bp = plt.boxplot([barcelona, madrid,revs], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')


plt.show()
```
